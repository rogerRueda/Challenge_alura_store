### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



En esta primera sección, calculamos cuánto dinero generó cada tienda en total. Sumamos todos los precios de los productos vendidos para tener una idea clara del volumen de ventas. Esto nos permite identificar cuál tienda tuvo mayores ingresos.

```python
ingreso_total_tienda1 = tienda['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()

print(f"Ingreso tienda 1: ${ingreso_total_tienda1:.2f}")
print(f"Ingreso tienda 2: ${ingreso_total_tienda2:.2f}")
print(f"Ingreso tienda 3: ${ingreso_total_tienda3:.2f}")
print(f"Ingreso tienda 4: ${ingreso_total_tienda4:.2f}")
```

```python
import matplotlib.pyplot as plt


tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]
colores = ['#FF6347', '#4682B4', '#32CD32', '#FFD700']

plt.figure(figsize=(8, 5))
bars = plt.bar(tiendas, ingresos, color=colores)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 1000, f'${yval:,.2f}', ha='center', va='bottom', fontsize=10)

plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 2. Ventas por categoría

Aquí analizamos cuántos productos se vendieron en cada categoría (por ejemplo, ropa, electrónica, accesorios, etc.). Esto nos ayuda a saber cuáles tipos de productos son los más populares en cada tienda y entender mejor el comportamiento de los clientes.

```python
prod_vendidos_tienda1 = tienda['Categoría del Producto'].value_counts()
prod_vendidos_tienda2 = tienda2['Categoría del Producto'].value_counts()
prod_vendidos_tienda3 = tienda3['Categoría del Producto'].value_counts()
prod_vendidos_tienda4 = tienda4['Categoría del Producto'].value_counts()

print("Ventas por categoría en la tienda 1:")
print(prod_vendidos_tienda1)
print("\nVentas por categoría en la tienda 2:")
print(prod_vendidos_tienda2)
print("\nVentas por categoría en la tienda 3:")
print(prod_vendidos_tienda3)
print("\nVentas por categoría en la tienda 4:")
print(prod_vendidos_tienda4)
```

```python
categorias_tienda1 = prod_vendidos_tienda1.index
categorias_tienda2 = prod_vendidos_tienda2.index
categorias_tienda3 = prod_vendidos_tienda3.index
categorias_tienda4 = prod_vendidos_tienda4.index

ventas_tienda1 = prod_vendidos_tienda1.values
ventas_tienda2 = prod_vendidos_tienda2.values
ventas_tienda3 = prod_vendidos_tienda3.values
ventas_tienda4 = prod_vendidos_tienda4.values

# Colores para cada tienda
colores_tiendas = ['#FF6347', '#4682B4', '#32CD32', '#FFD700']  # Rojo, Azul, Verde, Amarillo


fig, axes = plt.subplots(2, 2, figsize=(14, 10))

bars_tienda1 = axes[0, 0].bar(categorias_tienda1, ventas_tienda1, color=colores_tiendas[0])
axes[0, 0].set_title('Ventas por Categoría - Tienda 1')
axes[0, 0].set_xlabel('Categorías')
axes[0, 0].set_ylabel('Cantidad Vendida')
axes[0, 0].tick_params(axis='x', rotation=45)

for bar in bars_tienda1:
    yval = bar.get_height()
    axes[0, 0].text(bar.get_x() + bar.get_width() / 2, yval + 0.5, f'{yval}', ha='center', va='bottom')


bars_tienda2 = axes[0, 1].bar(categorias_tienda2, ventas_tienda2, color=colores_tiendas[1])
axes[0, 1].set_title('Ventas por Categoría - Tienda 2')
axes[0, 1].set_xlabel('Categorías')
axes[0, 1].set_ylabel('Cantidad Vendida')
axes[0, 1].tick_params(axis='x', rotation=45)

for bar in bars_tienda2:
    yval = bar.get_height()
    axes[0, 1].text(bar.get_x() + bar.get_width() / 2, yval + 0.5, f'{yval}', ha='center', va='bottom')

bars_tienda3 = axes[1, 0].bar(categorias_tienda3, ventas_tienda3, color=colores_tiendas[2])
axes[1, 0].set_title('Ventas por Categoría - Tienda 3')
axes[1, 0].set_xlabel('Categorías')
axes[1, 0].set_ylabel('Cantidad Vendida')
axes[1, 0].tick_params(axis='x', rotation=45)

for bar in bars_tienda3:
    yval = bar.get_height()
    axes[1, 0].text(bar.get_x() + bar.get_width() / 2, yval + 0.5, f'{yval}', ha='center', va='bottom')


bars_tienda4 = axes[1, 1].bar(categorias_tienda4, ventas_tienda4, color=colores_tiendas[3])
axes[1, 1].set_title('Ventas por Categoría - Tienda 4')
axes[1, 1].set_xlabel('Categorías')
axes[1, 1].set_ylabel('Cantidad Vendida')
axes[1, 1].tick_params(axis='x', rotation=45)

for bar in bars_tienda4:
    yval = bar.get_height()
    axes[1, 1].text(bar.get_x() + bar.get_width() / 2, yval + 0.5, f'{yval}', ha='center', va='bottom')


plt.tight_layout()


plt.show()
```

# 3. Calificación promedio de la tienda


En este apartado, calculamos la calificación promedio que los clientes le dieron a los productos en cada tienda. Una calificación alta indica que los clientes están satisfechos, mientras que una baja puede señalar áreas de mejora en la calidad de los productos o el servicio.

```python
calificacion_promedio1 = tienda['Calificación'].mean()
calificacion_promedio2 = tienda2['Calificación'].mean()
calificacion_promedio3 = tienda3['Calificación'].mean()
calificacion_promedio4 = tienda4['Calificación'].mean()

print(f"Calificación promedio tienda 1: {calificacion_promedio1:.2f}")
print(f"Calificación promedio tienda 2: {calificacion_promedio2:.2f}")
print(f"Calificación promedio tienda 3: {calificacion_promedio3:.2f}")
print(f"Calificación promedio tienda 4: {calificacion_promedio4:.2f}")
```

```python
calificaciones = [calificacion_promedio1, calificacion_promedio2, calificacion_promedio3, calificacion_promedio4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
colores_tiendas = ['#FF6347', '#4682B4', '#32CD32', '#FFD700']

plt.figure(figsize=(8, 5))
bars = plt.barh(tiendas, calificaciones, color=colores_tiendas)


for bar in bars:
    plt.text(bar.get_width() + 0.02, bar.get_y() + bar.get_height() / 2, f'{bar.get_width():.2f}',
             va='center', ha='left', fontsize=10)


plt.title('Calificación Promedio por Tienda')
plt.xlabel('Calificación Promedio')
plt.ylabel('Tiendas')
plt.grid(axis='x', linestyle='--', alpha=0.7)


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [1150880400, 1116343500, 1098019600, 1038375700]
calificaciones = [3.98, 4.04, 4.05, 4.00]
envios = [26018.61, 25216.24, 24805.68, 23459.46]


x = np.arange(len(tiendas))
width = 0.25


fig, ax1 = plt.subplots(figsize=(12, 6))


bar1 = ax1.bar(x - width, ingresos, width, label='Ingresos', color='#FF6347')

bar2 = ax1.bar(x, [c * 1e8 for c in calificaciones], width, label='Calificación Promedio (x1e8)', color='#4682B4')

bar3 = ax1.bar(x + width, [e * 40000 for e in envios], width, label='Costo Envío Promedio (x40,000)', color='#32CD32')


ax1.set_xlabel('Tiendas')
ax1.set_ylabel('Valores')
ax1.set_title('Comparativa entre Tiendas')
ax1.set_xticks(x)
ax1.set_xticklabels(tiendas)
ax1.legend()

plt.tight_layout()
plt.show()
```
